```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.linear_model import LinearRegression
```

```python
n_samples = 1000
player_age = np.random.choice(22, n_samples)+18
player_weight = np.random.choice(90, n_samples)+60
player_height = np.random.choice(60, n_samples)+160
goals_assists = np.random.choice(120, n_samples)*((100-player_age)/100)
years_in_NHL = np.random.choice(player_age, n_samples)-17


```

###### Создаем датасет с данными игроков НХЛ

```python
data = pd.DataFrame({'age':player_age, 'height':player_height, 'weight':player_weight, 'average per season':goals_assists, 
                     'years in NHL':years_in_NHL})
data
```

```python
from sklearn.metrics import mean_absolute_error
```

###### Целевой переменной будет "средняя статистика" в сезон

```python
X = data[['age', 'height', 'weight', 'years in NHL']]
y = data['average per season']

reg = LinearRegression().fit(X,y)

print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))
pred_values = reg.predict(data[['age', 'height', 'weight', 'years in NHL']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```

```python
data['career points total']= data['average per season'] * (data['years in NHL']+(40-data['years in NHL']))
data
```

###### Предсказываем общее количесвто очков за карьеру если играет до 40 лет

```python
X = data[['career points total']]
y = data['average per season']
reg = LinearRegression().fit(X, y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(data[['career points total']])
print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```
